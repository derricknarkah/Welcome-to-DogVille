# Welcome to DogVille

"That is a doggo! No! it looks more like a pupper."
"I will go for floofer eventhough my initial guess would have been a puppa".

Do not try reading the text above again. It only takes a single read for me to tell if you are part of the committed league of domestic dog lovers or on the other side of the wall. I discovered this large community who took exceptional interest in dogs and even had a twitter platform where images of these pets could be rated. Wow!.

As part of a project aimed at sharpening my skill in Data wrangling, I took a gently stroll through dogville. I extracted a couple of tweets from the twitter user @dog_rages also known as WeRateDogs. This account has made a reputation out of rating images of dog tweets with a unique rating system.  The data collected in conjuction with data provided by the account administrators when assessed, cleaned and analysed revealed a few things about dogs. 

The clean data had about 2000 entries. Each entry showed the tweets, the WeRateDogs rating, the name of the dog and the stage of the dog. Right, stage of the dog. Apparently, dogs either fell into categories like doggo, floofer, pupper, puppa and the like. You should quickly check these out.

Now at a first glance, the total number of likes for tweets about dogs in the pupper stage was the highest (1226358) and could easily make one assume pupper images were the most lilked. Not soo fast! There were also more pupper-related tweets than the rest.(About twice the number of tweets concerning the other dogs.) A rather careful look at the average likes a pupper tweet received told a different story. The mean was lower than the rest. A pupper related tweet got about only 5625 likes. Maybe they are simply the most common.

The second thing I noticed was that the puppo staged dogs seemed to be the most liked. These had the highest average likes of (18683). Even from the data, the most liked tweet was a puppo tweet. Are they simply adorable? Were the tweeted images taken by professional photographers? For such a few number of tweets getting the most average likes painted a clear picture of the interests of the lot.

Lastly, there was an observable correlation between the number of retweets and the number of likes of the dogville posts. The purpose of a retweet or a like is definitely a discretional matter but at least we know one thing about our dog lovers. Their opinions can easily be seen through either retweets and likes.
